# Pokemons que são ruins contra água

```python
weak_against_water_df = pokemon_df[pokemon_df['against_water'] > 1]

weak_against_water_df[['english_name', 'primary_type', 'against_water']]
```

# Pokemons muito ruins contra água

```python
weak_against_water_df = pokemon_df[pokemon_df['against_water'] > 2]

weak_against_water_df[['english_name', 'primary_type', 'against_water']]
```

# Pokemons que são fracos contra água e terra ao mesmo tempo


```python
weak_against_water_df = pokemon_df[(pokemon_df['against_water'] > 1) & (pokemon_df['against_ground'] > 1)]
weak_against_water_df [['english_name', 'against_water', 'against_ground']]
```
